```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import requests
import os
import json
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import WebDriverException
from selenium.common.exceptions import NoSuchElementException
```

```python
model_name = [] #done
model_dis_price = [] #done
model_org_price = [] #done
model_flip_assured = [] #done
model_rating = [] #done
model_rating_review_count = [] #done
model_processor = [] #done
model_ram = [] #done
model_op_system = [] #done
model_display = [] #done
model_warranty = [] #done

PATH = r'C:\Users\FQ2071TU\chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36")
options.headless = True  
driver = webdriver.Chrome(options=options, executable_path=PATH)

web_base_url = 'https://www.flipkart.com/search?q=laptop+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D50000&page='
for page_num in range(1,19):
    web_pages=page_num
    concat_web_url= web_base_url + str(web_pages)
    

    driver.get(concat_web_url)
    time.sleep(4)
    products = driver.find_elements(By.XPATH, "//div[@class='_13oc-S']")
    for product in products:
        try:
            if_ad = product.find_element(By.XPATH, ".//div[@class='_2tfzpE']")
            continue
        except NoSuchElementException:
            try:
                name = product.find_element(By.XPATH, ".//div[@class='_4rR01T']")
                model_name.append(name.text)
            except NoSuchElementException:
                model_name.append('NA')
            try:
                dis_price = product.find_element(By.XPATH, ".//div[@class='_30jeq3 _1_WHN1']")
                model_dis_price.append(dis_price.text)
            except NoSuchElementException:
                model_dis_price.append('NA')

            try:
                org_price = product.find_element(By.XPATH, ".//div[@class='_3I9_wc _27UcVY']")
                model_org_price.append(org_price.text)
            except NoSuchElementException:
                model_org_price.append('NA')
                

            try:
                flip_assured = product.find_element(By.XPATH, "//div[@class='_13J9qT']")
                model_flip_assured.append('Yes')
            except NoSuchElementException:
                model_flip_assured.append('No')

            try:
                ratings = product.find_element(By.XPATH, ".//div[@class='_3LWZlK']")
                model_rating.append(ratings.text)
            except NoSuchElementException:
                model_rating.append('NA')

            try:
                rating_review_count = product.find_element(By.XPATH, ".//span[@class='_2_R_DZ']")
                model_rating_review_count.append(rating_review_count.text)
            except NoSuchElementException:
                model_rating_review_count.append('NA')

            try:
                desc = product.find_element(By.XPATH, ".//ul[@class='_1xgFaf']")
                desc_lines = desc.text.split('\n')
                for line in desc_lines:
                    if 'display' in line.lower():
                        display = line
                        model_display.append(display)
                        display_found = True
                        break
                if not display_found:
                    model_display.append('NA')
                
                for line in desc_lines:
                    if 'warranty' in line.lower():
                        warranty = line
                        model_warranty.append(warranty)
                        warranty_found = True
                        break
                if not warranty_found:
                    model_warranty.append('NA')
                        
                for line in desc_lines:
                    if 'processor' in line.lower():
                        processor = line
                        model_processor.append(processor)
                        processor_found = True
                        break
                if not processor_found:
                    model_processor.append('NA')
                        
                for line in desc_lines:
                    if 'operating' in line.lower():
                        operating = line
                        model_op_system.append(operating)
                        operating_found = True
                        break
                if not operating_found:
                    model_op_system.append('NA')
                        
                for line in desc_lines:
                    if 'ram' in line.lower():
                        ram = line
                        model_ram.append(ram)
                        ram_found = True
                        break
                if not ram_found:
                    model_ram.append('NA')

            except NoSuchElementException:
                print('no DESCRIPTION') 
    print(page_num)
    print('***')

    
```
